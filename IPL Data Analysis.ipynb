## IPL Dataset

```python
# read data
data=pd.read_csv('D:\datasets\IPL Matches 2008-2020.csv')
data.head()
```

```python
df=pd.DataFrame(data['date'])
df[["month", "day", "year"]]=df["date"].str.split("/", expand = True)
```

```python
df.head()
```

```python
data.drop(labels='date',axis=1,inplace=True)
data.insert(2,'Year',df['year'])
data.head()
```

```python
data['Year']=data['Year'].astype(int)
```

### In 2008-2015 how many matches played by Virat Kohli and  Rohit Sharma also while playing matches in 2008-2015 how many matches they win?

```python
df1=data.loc[((data['Year']>=2008)&(data['Year']<=2015))]
```

```python
df1.head()
```

```python
matches_played=df1.loc[((data['player_of_match']=='R Sharma')|(data['player_of_match']=='V Kohli'))]
matches_played.head()
```

```python
matches_played['player_of_match'].value_counts()
    
```
